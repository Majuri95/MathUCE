#üßÆ Funciones: Utilidad, Tipos y Composici√≥n
## üìòIntroducci√≥n:

Las funciones son herramientas matem√°ticas fundamentales que permiten modelar relaciones entre variables y son esenciales en diversos campos del conocimiento. En este an√°lisis detallado, exploraremos la utilidad, tipos y composici√≥n de las funciones, proporcionando una comprensi√≥n profunda de estos conceptos.
##üîçUtilidad de las funciones:

Las funciones ofrecen una amplia gama de beneficios en la resoluci√≥n de problemas y el an√°lisis de datos:

* Modelado de relaciones: Permiten representar la relaci√≥n entre dos conjuntos de n√∫meros, facilitando la comprensi√≥n de fen√≥menos y comportamientos.

* Simplificaci√≥n de c√°lculos: Facilitan la resoluci√≥n de problemas complejos al encapsular operaciones repetitivas en una √∫nica expresi√≥n.

* Generalizaci√≥n de soluciones: Permiten encontrar soluciones generales a problemas similares, sin necesidad de repetir c√°lculos para cada caso espec√≠fico.

* An√°lisis de cambio: Facilitan el estudio de c√≥mo cambia el valor de salida a medida que var√≠a el valor de entrada, proporcionando informaci√≥n sobre la sensibilidad de la funci√≥n.

* Comunicaci√≥n matem√°tica: Brindan una forma concisa y precisa de expresar relaciones entre variables, mejorando la comunicaci√≥n y el intercambio de ideas.

## üë®‚Äçüíª Definici√≥n y Representaci√≥n de Funciones

Una funci√≥n es una regla de asociaci√≥n que asigna a cada elemento de un conjunto de partida (A) un elemento √∫nico en un conjunto de llegada (B). La notaci√≥n matem√°tica est√°ndar para una funci√≥n es $(f: A \rightarrow B$). Visualmente, podemos imaginar esto como un mapa que toma cada elemento $(x$) en $(A$fv) y lo asocia con un elemento $(f(x)$) en $(B$).


Imaginemos que $(A$) es un conjunto de n√∫meros y $(B$) es un conjunto de colores. La funci√≥n $(f$) podr√≠a representar una regla que asocia cada n√∫mero con un color espec√≠fico.

##üß©Tipos de funciones:

Existe una amplia variedad de funciones, cada una con caracter√≠sticas y aplicaciones espec√≠ficas:

* Funci√≥n lineal: Representada por y = mx + b, donde m y b son constantes. Su gr√°fica es una l√≠nea recta.
![Texto alternativo](https://upload.wikimedia.org/wikipedia/commons/4/4c/Funci%C3%B3n_lineal.svg)

* Funci√≥n cuadr√°tica: Representada por y = ax^2 + bx + c, donde a, b y c son constantes. Su gr√°fica es una par√°bola.

![Texto alternativo](https://upload.wikimedia.org/wikipedia/commons/f/f8/Polynomialdeg2.svg)

* Funci√≥n exponencial: Representada por y = a^x, donde a es una constante positiva (a ‚â† 1). Su gr√°fica muestra un crecimiento o decrecimiento exponencial.

![Texto alternativo](https://www.fisicalab.com/sites/all/files/contenidos/matematicas/9097_f_exponenciales/graficas_exponenciales.jpg)

* Funci√≥n logar√≠tmica: Representada por y = log_a(x), donde a es una constante positiva (a ‚â† 1). Su gr√°fica es la inversa de la funci√≥n exponencial.

![Texto alternativo](https://www.fisicalab.com/sites/all/files/contenidos/matematicas/9209_f_logaritmicas/otras_graficas_logaritmicas.jpg)

* Funci√≥n trigonom√©trica: Representada por y = sin(x), y = cos(x), y = tan(x). Sus gr√°ficas muestran comportamientos peri√≥dicos.

![Texto alternativo](https://www.varsitytutors.com/assets/vt-hotmath-legacy/hotmath_help/topics/trigonometic-functions/sine.gif)

* Funci√≥n compuesta: Se obtiene al combinar dos o m√°s funciones, aplicando una funci√≥n al resultado de otra.

![Texto alternativo](https://www.neurochispas.com/wp-content/uploads/2022/08/diagrama-de-mapeo-de-una-funcion-compuesta-g-of-f-600x281.png)

* Funci√≥n inversa: Representa la funci√≥n que "deshace" lo que hace la funci√≥n original. Si f(x) = y, entonces f^(-1)(y) = x.

![Texto alternativo](https://www.fisicalab.com/sites/all/files/contenidos/matematicas/2225_f_inversa/ejemplos_inversas.jpg)

### Ejemplos


```python
import math
import matplotlib.pyplot as plt

# Funci√≥n lineal
def funcion_lineal(x, m, b):
  """
  Funci√≥n lineal y = mx + b.

  Args:
    x: Valor de entrada (variable independiente).
    m: Pendiente de la recta.
    b: Intercepto en el eje y.

  Returns:
    Valor de salida (imagen) de la funci√≥n lineal para el valor de entrada x.
  """
  return m * x + b

# Funci√≥n cuadr√°tica
def funcion_cuadratica(x, a, b, c):
  """
  Funci√≥n cuadr√°tica y = ax^2 + bx + c.

  Args:
    x: Valor de entrada (variable independiente).
    a: Coeficiente cuadr√°tico.
    b: Coeficiente lineal.
    c: T√©rmino independiente.

  Returns:
    Valor de salida (imagen) de la funci√≥n cuadr√°tica para el valor de entrada x.
  """
  return a * x**2 + b * x + c

# Funci√≥n exponencial
def funcion_exponencial(x, a):
  """
  Funci√≥n exponencial y = a^x.

  Args:
    x: Valor de entrada (variable independiente).
    a: Base de la exponencial (a > 0, a ‚â† 1).

  Returns:
    Valor de salida (imagen) de la funci√≥n exponencial para el valor de entrada x.
  """
  return a**x

# Funci√≥n logar√≠tmica
def funcion_logaritmica(x, a):
  """
  Funci√≥n logar√≠tmica y = log_a(x).

  Args:
    x: Valor de entrada (variable independiente) (x > 0).
    a: Base de la logaritmo (a > 0, a ‚â† 1).

  Returns:
    Valor de salida (imagen) de la funci√≥n logar√≠tmica para el valor de entrada x.
  """
  return math.log(x, a)

# Funci√≥n trigonom√©trica seno
def funcion_trigonometrica_sin(x):
  """
  Funci√≥n trigonom√©trica seno y = sin(x).

  Args:
    x: Valor de entrada (variable independiente) en radianes.

  Returns:
    Valor de salida (imagen) de la funci√≥n seno para el valor de entrada x.
  """
  return math.sin(x)

# Funci√≥n trigonom√©trica coseno
def funcion_trigonometrica_cos(x):
  """
  Funci√≥n trigonom√©trica coseno y = cos(x).

  Args:
    x: Valor de entrada (variable independiente) en radianes.

  Returns:
    Valor de salida (imagen) de la funci√≥n coseno para el valor de entrada x.
  """
  return math.cos(x)

# Funci√≥n trigonom√©trica tangente
def funcion_trigonometrica_tan(x):
  """
  Funci√≥n trigonom√©trica tangente y = tan(x).

  Args:
    x: Valor de entrada (variable independiente) en radianes.

  Returns:
    Valor de salida (imagen) de la funci√≥n tangente para el valor de entrada x.
  """
  return math.tan(x)

# Funci√≥n inversa lineal
def funcion_inversa_lineal(y, m, b):
  """
  Funci√≥n inversa de la funci√≥n lineal y = mx + b.

  Args:
    y: Valor de salida (imagen).
    m: Pendiente de la recta.
    b: Intercepto en el eje y.

  Returns:
    Valor de entrada (preimagen) x que satisface y = mx + b.
  """
  return (y - b) / m


```

## üß©Funci√≥n Sobreyectiva, Biyectiva, Inyectiva, Inversa y Compuesta

#### üåàFunci√≥n Sobreyectiva (Suprayectiva o Exhaustiva):

Una funci√≥n f: X ‚Üí Y se considera sobreyectiva si para cada elemento y perteneciente al codominio Y, existe al menos un elemento x del dominio X tal que f(x) = y. En otras palabras, la imagen de la funci√≥n (el conjunto de valores que toma f(x) para todos los x en el dominio) cubre todo el codominio.

####Representaci√≥n gr√°fica:

En una gr√°fica de una funci√≥n sobreyectiva, cada punto del codominio Y se encuentra alineado con al menos un punto del dominio X.


#### Ejemplos de funciones sobreyectivas:

La funci√≥n f(x) = x^2 + 1, donde X es el conjunto de todos los n√∫meros reales y Y es el conjunto de n√∫meros mayores o iguales a 1.
La funci√≥n g(x) = 2x + 1, donde X es el conjunto de todos los n√∫meros reales y Y es el conjunto de todos los n√∫meros reales.
Funci√≥n Biyectiva (Inyectiva y Sobreyectiva a la vez):

Una funci√≥n f: X ‚Üí Y se considera biyectiva si cumple dos condiciones:

### üîÑFunci√≥n Biyectividad:

Para cualquier par de elementos distintos x1 y x2 en el dominio X, sus im√°genes f(x1) y f(x2) tambi√©n son distintas en el codominio Y. En otras palabras, la funci√≥n no asigna el mismo valor de salida a dos elementos de entrada diferentes.

Sobreyectividad: La funci√≥n f es sobreyectiva, es decir, cubre todo el codominio Y.

####Representaci√≥n gr√°fica:

En una gr√°fica de una funci√≥n biyectiva, cada punto del codominio Y se encuentra alineado con exactamente un punto del dominio X, y viceversa. La gr√°fica de una funci√≥n biyectiva es una recta diagonal.

####Ejemplos de funciones biyectivas:

La funci√≥n f(x) = x, donde X e Y son el conjunto de todos los n√∫meros reales.
La funci√≥n g(x) = 2x - 1, donde X es el conjunto de todos los n√∫meros reales y Y es el conjunto de todos los n√∫meros reales excepto 1/2.

###üìäFunci√≥n Inyectiva (Univalente):

Una funci√≥n f: X ‚Üí Y se considera inyectiva si para cualquier par de elementos distintos x1 y x2 en el dominio X, sus im√°genes f(x1) y f(x2) tambi√©n son distintas en el codominio Y. En otras palabras, la funci√≥n no asigna el mismo valor de salida a dos elementos de entrada diferentes.

####Representaci√≥n gr√°fica:

En una gr√°fica de una funci√≥n inyectiva, cada punto del codominio Y se encuentra alineado con a lo sumo un punto del dominio X.

####Ejemplos de funciones inyectivas:

La funci√≥n f(x) = x^2, donde X es el conjunto de todos los n√∫meros reales y Y es el conjunto de todos los n√∫meros mayores o iguales a 0.
La funci√≥n g(x) = e^x, donde X es el conjunto de todos los n√∫meros reales y Y es el conjunto de todos los n√∫meros reales positivos.
Funci√≥n Inversa:

Para una funci√≥n f: X ‚Üí Y que sea biyectiva, existe una funci√≥n inversa f^(-1): Y ‚Üí X tal que para cualquier elemento x en el dominio X y su imagen y en el codominio Y, se cumple que f^(-1)(y) = x y f(f^(-1)(y)) = y. En otras palabras, la funci√≥n inversa "deshace" lo que hace la funci√≥n original.

### üìä Propiedades de la funci√≥n inversa:

La funci√≥n inversa de una funci√≥n biyectiva es tambi√©n biyectiva.
La gr√°fica de la funci√≥n inversa es la sim√©trica de la gr√°fica de la funci√≥n original respecto a la bisectriz de los cuadrantes.
Ejemplo de funci√≥n inversa:

La funci√≥n f(x) = x^2, donde X es el conjunto de todos los n√∫meros reales no negativos y Y es el conjunto de todos los n√∫meros reales no negativos, tiene una funci√≥n inversa f^(-1

```python
import matplotlib.pyplot as plt
import numpy as np

# Definici√≥n de funciones
def f_inyectiva(x):
    return x

def f_sobreyectiva(x):
    return np.tanh(x)

def f_biyectiva(x):
    return x**2

def f_inversa(x):
    return np.sqrt(x)

# Valores de x para graficar
x = np.linspace(-1, 1, 200)

# Crear la figura
plt.figure(figsize=(6, 4))

# Funci√≥n Inyectiva
plt.subplot(131)
plt.plot(x, f_inyectiva(x), label='$f(x) = x$', color='blue')
plt.title('Funci√≥n Inyectiva')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Funci√≥n Sobreyectiva
plt.subplot(132)
plt.plot(x, f_sobreyectiva(x), label='$f(x) = tanh(x)$', color='orange')
plt.title('Funci√≥n Sobreyectiva')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Funci√≥n Biyectiva
plt.subplot(133)
plt.plot(x, f_biyectiva(x), label='$f(x) = x^3$', color='green')
plt.title('Funci√≥n Biyectiva')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Funci√≥n Inversa

plt.plot(x, f_inversa(x), label='$f^{-1}(x) = \sqrt{x}$', color='red')
plt.title('Funci√≥n Inversa')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Ajustar el dise√±o de la figura
plt.tight_layout()

# Mostrar la figura
plt.show()
```

###üö∞ Funci√≥n de Utilidad

Consideremos un ejemplo simple en el que el √∫nico bien disponible es agua. La funci√≥n de utilidad, en este caso, puede representarse como $( u(x) $), donde $( x $) es la cantidad de agua consumida y \( u(x) \) es la utilidad obtenida.

La utilidad marginalmente decrece, lo que significa que cada unidad adicional de agua proporciona una utilidad menor que la anterior.



```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definimos una funci√≥n de utilidad diferente
def u(x, y):
    return np.sin(x) * np.cos(y)

# Rango de valores para x e y
x = np.linspace(-2, 2, 200)
y = np.linspace(-2, 2, 200)
X, Y = np.meshgrid(x, y)
Z = u(X, Y)

# Creamos la figura
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111, projection='3d')

# Gr√°fico de la superficie de la funci√≥n de utilidad
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Contornos en el plano
contours = ax.contour(X, Y, Z, levels=np.linspace(-1, 1, 10), cmap='autumn', linestyles="solid")

# Etiquetas y t√≠tulo
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Utilidad U(x, y)')
ax.set_title('Superficie de la Funci√≥n de Utilidad y Contornos')

# Barra de color
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

# Mostrar el gr√°fico
plt.show()
```

   ]$
   Esta notaci√≥n indica que la funci√≥n de utilidad \( U \) toma como entrada un par de n√∫meros reales no negativos $(( \mathbb{R}^2_+ ))$ y produce un n√∫mero real no negativo $(( \mathbb{R}^+ ))$.

2. **Expresi√≥n de la Funci√≥n de Utilidad**:
   $[
   (x, y) \rightarrow u(x, y)
   ]$
