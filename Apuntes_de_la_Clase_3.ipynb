# üåü Sistemas Num√©ricos

Los sistemas num√©ricos son m√©todos para representar n√∫meros. Python soporta varios sistemas num√©ricos, los m√°s comunes son el decimal, binario, octal y hexadecimal.

1. Decimal (Base 10): Es el sistema num√©rico m√°s utilizado, que usa los d√≠gitos del 0 al 9. Por ejemplo, 123 en decimal es simplemente 123.


2. Binario (Base 2): Utiliza solo los d√≠gitos 0 y 1. Es el sistema num√©rico fundamental en la inform√°tica y la electr√≥nica digital. En Python, los n√∫meros binarios se representan con el prefijo 0b o 0B. Por ejemplo:

```python
binario = 0b1010  # Representa el n√∫mero 10 en decimal
```

3. Octal (Base 8): Utiliza los d√≠gitos del 0 al 7. En Python, los n√∫meros octales se representan con el prefijo 0o o 0O. Por ejemplo:

```python
octal = 0o12  # Representa el n√∫mero 10 en decimal
```

 4. Hexadecimal (Base 16): Utiliza los d√≠gitos del 0 al 9 y las letras A a F (o a-f) para representar valores del 10 al 15. En Python, los n√∫meros hexadecimales se representan con el prefijo 0x o 0X. Por ejemplo:

```python
hexadecimal = 0xA  # Representa el n√∫mero 10 en decimal
```

Python proporciona funciones integradas para convertir entre estos sistemas num√©ricos:

bin() para convertir a binario.

oct() para convertir a octal.

hex() para convertir a hexadecimal.

Ejemplo:

```python
numero = 10
print(bin(numero))  # Output: 0b1010
print(oct(numero))  # Output: 0o12
print(hex(numero))  # Output: 0xA
```

#üìäCardinalidad
##üìè Qu√© es Cardinalidad?

En el √°mbito de las matem√°ticas, la cardinalidad, tambi√©n conocida como n√∫mero cardinal, se refiere a la medida de la "cantidad de elementos" que conforman un conjunto.

En palabras m√°s simples, la cardinalidad nos indica cu√°ntos elementos tiene un conjunto. Se representa com√∫nmente con el s√≠mbolo "|A|", donde A es el nombre del conjunto en cuesti√≥n.

Para conjuntos finitos, como por ejemplo A = {2, 4, 6}, la cardinalidad se calcula simplemente contando el n√∫mero de elementos. En este caso, el conjunto A tiene 3 elementos, por lo que su cardinalidad ser√≠a |A| = 3.

Sin embargo, la cardinalidad tambi√©n se aplica a conjuntos infinitos, donde contar los elementos de forma individual no es posible. En estos casos, se utilizan conceptos m√°s avanzados como los n√∫meros cardinales infinitos.

##üî¢Ejemplo de cardinalidad en conjuntos infinitos:

El conjunto de los n√∫meros naturales (N): N = {1, 2, 3, ...}, tiene una cardinalidad infinita denominada aleph-null (‚Ñµ‚ÇÄ).

El conjunto de los n√∫meros enteros (Z): Z = {..., -2, -1, 0, 1, 2, ...}, tambi√©n tiene cardinalidad aleph-null (‚Ñµ‚ÇÄ), ya que contiene a los n√∫meros naturales como subconjunto.

En el contexto de las bases de datos:

La cardinalidad se utiliza para describir la relaci√≥n entre dos entidades en un modelo de datos. Indica cu√°ntas veces una instancia de una entidad puede estar relacionada con instancias de otra entidad.

##üìöTipos de cardinalidad en bases de datos:

Uno a uno (1:1): Una instancia de una entidad puede estar relacionada con m√°ximo una instancia de otra entidad.

Uno a varios (1:N): Una instancia de una entidad puede estar relacionada con varias instancias de otra entidad.

Varios a varios (N:N): Una instancia de una entidad puede estar relacionada con varias instancias de otra entidad, y viceversa.

### üìä Cardinalidad de Conjuntos

La **cardinalidad** de un conjunto es una medida del "n√∫mero de elementos" en el conjunto. Los conjuntos pueden ser finitos o infinitos. La cardinalidad se representa com√∫nmente como |A|, donde A es el conjunto.

Ejemplos:
1. N√∫meros Aleatorios
- Si A = {1, 2, 3}, entonces |A| = 3.
- Si B = {x ‚àà ‚Ñï | x < 10}, entonces |B| = 9.



```python
# Ejemplo de cardinalidad
A = {1, 2, 3}
B = set(range(10))

print(f"Cardinalidad de A: {len(A)}")
print(f"Cardinalidad de B: {len(B)}")
```

2. Conjunto de n√∫meros pares:

```python
# Definir el conjunto de n√∫meros pares
pares = {2, 4, 6, 8, 10}

# Calcular la cardinalidad (n√∫mero de elementos) del conjunto
cardinalidad_pares = len(pares)

# Imprimir la cardinalidad
print("Cardinalidad del conjunto de n√∫meros pares:", cardinalidad_pares)
```

3. Uni√≥n de dos conjuntos:

```python
# Definir dos conjuntos
conjunto_A = {1, 2, 3}
conjunto_B = {3, 4, 5}

# Calcular la uni√≥n de los conjuntos (A U B)
union_AB = conjunto_A | conjunto_B

# Calcular la cardinalidad de la uni√≥n
cardinalidad_union = len(union_AB)

# Imprimir la cardinalidad
print("Cardinalidad de la uni√≥n (A U B):", cardinalidad_union)
```

```python
# Generaci√≥n de un Diagrama de Venn para dos conjuntos
from matplotlib_venn import venn2
import matplotlib.pyplot as plt # Import the matplotlib.pyplot module

# Conjuntos de Ejemplo
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

# Diagrama de Venn
plt.figure(figsize=(8, 8))
venn2([A, B], ('Conjunto A', 'Conjunto B'))
plt.title("Diagrama de Venn de los Conjuntos A y B")
plt.show()
```
