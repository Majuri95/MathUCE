# 📊Funciones Matemáticas: Imagen y Preimagen
## 🔍 Qué son las funciones matemáticas?

En el mundo de las matemáticas, las funciones son conceptos fundamentales que describen la relación entre dos conjuntos de números. Se las puede visualizar como reglas o mecanismos que asignan un valor de salida (llamado imagen) a cada valor de entrada (llamado dominio).

![Texto alternativo](https://datascientest.com/es/wp-content/uploads/sites/7/2022/06/ejemplo_fonciones-1-1024x690.png)


### 📘Ejemplo

Imaginemos una máquina expendedora:

* Entrada: Insertas una moneda (valor de entrada).
Función: La máquina interpreta la moneda y determina el producto correspondiente (regla o mecanismo).
Salida: Recibes el producto seleccionado (imagen o valor de salida).
Características esenciales de las funciones:

* Dominio: El conjunto de todos los valores de entrada válidos para la función. Se representa con la letra D.

* Codominio: El conjunto que contiene todos los valores de salida posibles de la función. Se representa con la letra C.

* Imagen: El subconjunto del codominio que contiene todos los valores de salida efectivamente asignados por la función. Se representa con la letra I.

* Unicidad: Para cada valor de entrada en el dominio, solo hay un valor de salida correspondiente en la imagen.

## 🧮Representaciones comunes de funciones:

* Fórmulas matemáticas: La forma más común de expresar una función es mediante una fórmula que relaciona la variable de entrada (x) con la variable de salida (y). Por ejemplo, la función lineal y = mx + b, donde m y b son constantes.

* Tablas de valores: Se muestran pares de valores de entrada y salida, permitiendo observar la relación entre ellos.

* Gráficas: Representan visualmente la relación entre la variable de entrada (eje x) y la variable de salida (eje y). La curva de la gráfica indica cómo cambia el valor de salida a medida que varía la entrada.

### 🗽Ejemplos:

```python
import math

def funcion(x):
  """
  Función f(x) = x^2 + 1.

  Args:
    x: Valor de entrada (variable independiente).

  Returns:
    Valor de salida (imagen) de la función f(x) para el valor de entrada x.
  """
  return x**2 + 1

def imagen(x_min, x_max):
  """
  Calcula la imagen de la función f(x) en el intervalo [x_min, x_max].

  Args:
    x_min: Limite inferior del intervalo.
    x_max: Limite superior del intervalo.

  Returns:
    Lista de valores de imagen (f(x)) para cada valor de x en el intervalo.
  """
  valores_imagen = []
  for x in range(x_min, x_max + 1):
    valor_imagen = funcion(x)
    valores_imagen.append(valor_imagen)
  return valores_imagen

def preimagen(y):
  """
  Calcula la preimagen (valores de x) para un valor de imagen y de la función f(x).

  Args:
    y: Valor de imagen (f(x)) para el que se busca la preimagen.

  Returns:
    Lista de valores de x que satisfacen f(x) = y.
  """
  valores_preimagen = []
  for x in range(-10, 11):  # Ajustar el rango según la función
    valor_imagen = funcion(x)
    if valor_imagen == y:
      valores_preimagen.append(x)
  return valores_preimagen

# Ejemplo de uso
x_min = -3
x_max = 3
y = 5

valores_imagen = imagen(x_min, x_max)
valores_preimagen = preimagen(y)

print("Imagen de f(x) en [", x_min, ",", x_max, "]:", valores_imagen)
print("Preimagen de f(x) para y =", y, ":", valores_preimagen)
```

```python
import math
import matplotlib.pyplot as plt

def funcion(x):
  """
  Función f(x) = x^2 + 1.

  Args:
    x: Valor de entrada (variable independiente).

  Returns:
    Valor de salida (imagen) de la función f(x) para el valor de entrada x.
  """
  return x**2 + 1

def imagen(x_min, x_max, n_puntos):
  """
  Calcula y grafica la imagen de la función f(x) en el intervalo [x_min, x_max].

  Args:
    x_min: Limite inferior del intervalo.
    x_max: Limite superior del intervalo.
    n_puntos: Numero de puntos para calcular la imagen.

  Returns:
    Listas de valores de x (eje x) y y (eje y) para la imagen de la función.
  """
  x_valores = []
  y_valores = []
  for i in range(n_puntos):
    x = x_min + (i * (x_max - x_min)) / (n_puntos - 1)
    y = funcion(x)
    x_valores.append(x)
    y_valores.append(y)
  return x_valores, y_valores

def preimagen(y, x_min, x_max, n_puntos):
  """
  Calcula y grafica la preimagen (valores de x) para un valor de imagen y de la función f(x).

  Args:
    y: Valor de imagen (f(x)) para el que se busca la preimagen.
    x_min: Limite inferior del intervalo de búsqueda.
    x_max: Limite superior del intervalo de búsqueda.
    n_puntos: Numero de puntos para calcular la preimagen.

  Returns:
    Lista de valores de x (eje x) que satisfacen f(x) = y.
  """
  x_preimagen = []
  for i in range(n_puntos):
    x = x_min + (i * (x_max - x_min)) / (n_puntos - 1)
    valor_imagen = funcion(x)
    if abs(valor_imagen - y) < 1e-5:  # Tolerancia para la aproximación
      x_preimagen.append(x)
  return x_preimagen

# Ejemplo de uso
x_min = -3
x_max = 3
n_puntos = 100
y = 5

x_imagen, y_imagen = imagen(x_min, x_max, n_puntos)
x_preimagen = preimagen(y, x_min, x_max, n_puntos)

# Graficar imagen
plt.plot(x_imagen, y_imagen, label="Imagen f(x)")
plt.axhline(y=y, color="red", linestyle="--", label=f"Imagen fija y = {y}")

# Graficar preimagen (si existe)
if x_preimagen:
  for x in x_preimagen:
    plt.axvline(x=x, color="blue", linestyle="--", label=f"Preimagen x = {x}")

plt.xlim([x_min, x_max])
plt.ylim([0, y_imagen[-1] + 1])  # Ajustar rango de y según la gráfica
plt.xlabel("Eje X (x)")
plt.ylabel("Eje Y (y)")
plt.title("Imagen y Preimagen de f(x) = x^2 + 1")
plt.legend()
plt.grid(True)
plt.show()
```

### 🔄 Definición de Función
Una función $(f$) de un conjunto $(A$) a un conjunto $(B$) es una relación que asigna a cada elemento de $(A$) exactamente un elemento de $(B$). Se denota como $(f: A \to B$).


### 📆 Tipos de Funciones

### 📉Función original:

Una función original, también conocida como función directa o simplemente función, se define como una relación que asigna a cada elemento del dominio (conjunto de valores de entrada válidos) un único elemento del codominio (conjunto de valores de salida posibles).

### Representación de una función original:

Fórmulas matemáticas: La forma más común de expresar una función original es mediante una fórmula que relaciona la variable de entrada (x) con la variable de salida (y).

Por ejemplo, la función lineal y = mx + b, donde m y b son constantes.

Tablas de valores: Se muestran pares de valores de entrada (x) y salida (y), permitiendo observar la correspondencia entre ellos.

Gráficas: Representan visualmente la relación entre la variable de entrada (eje x) y la variable de salida (eje y).

 La curva de la gráfica indica cómo cambia el valor de salida a medida que varía la entrada.

###🔎Propiedades de una función original:

Unicidad: Para cada valor de entrada en el dominio, solo hay un valor de salida correspondiente en la imagen (el subconjunto del codominio que contiene los valores de salida efectivamente asignados).

No necesariamente invertible: No todas las funciones originales tienen una función inversa. Una función solo es invertible si es monótona y continua en su inverso.

### Ejemplos de funciones originales:

Función lineal: y = mx + b

Función cuadrática: y = ax^2 + bx + c

Función exponencial: y = a^x (donde a > 0, a ≠ 1)

Función logarítmica: y = log_a(x) (donde a > 0, a ≠ 1)

Función trigonométrica: y = sin(x), y = cos(x), y = tan(x)

##📒Función inversa:

La función inversa, también conocida como función recíproca, es la función que "deshace" lo que hace la función original. En otras palabras, si f(x) = y, entonces la función inversa f^(-1)(y) = x, donde x es el valor de entrada original que produce la salida y.

###Representación de una función inversa:

Fórmula: Si la función original tiene una fórmula explícita, la función inversa puede obtenerse despejando x en y = f(x).

Gráfica: La gráfica de la función inversa es la simétrica de la gráfica de la función original respecto a la bisectriz de los cuadrantes.
Propiedades de una función inversa:

Inversa de la original: Si f(f^(-1)(y)) = y para todo y en el dominio de la función inversa, y f^(-1)(f(x)) = x para todo x en el dominio de la función original, entonces f y f^(-1) son funciones inversas entre sí.

Monotonicidad: La función inversa es monótona (creciente o decreciente) si y solo si la función original es monótona estrictamente creciente o decreciente.

###Ejemplos de funciones inversas:

Función lineal inversa: y = (1/m) * (x - b)

Función cuadrática inversa: x = (-b ± sqrt(b^2 - 4ac)) / 2a

Función exponencial inversa: x = log_a(y) (donde a > 0, a ≠ 1)

Función logarítmica inversa: x = e^y

Función trigonométrica inversa: x = arcsin(y), x = arccos(y), x = arctan(y)

##👨‍💻Relación entre función original e inversa:

La gráfica de la función inversa se obtiene tomando la gráfica de la función original y reflejándola respecto a la bisectriz de los cuadrantes.

Las funciones inversas solo existen para funciones originales que son monótonas estrictamente crecientes o decrecientes en todo su dominio.

Si f(x) = y, entonces f^(-1)(y) = x, donde x es el valor de entrada original que produce la salida y.




```python
import math
import matplotlib.pyplot as plt

def funcion_original(x):
  """
  Función original f(x) = x^2 + 1.

  Args:
    x: Valor de entrada (variable independiente).

  Returns:
    Valor de salida (imagen) de la función f(x) para el valor de entrada x.
  """
  return x**2 + 1

def funcion_inversa(y):
  """
  Función inversa f^-1(y) de f(x) = x^2 + 1.

  Args:
    y: Valor de salida (imagen) para el que se busca la preimagen.

  Returns:
    Valor de entrada (preimagen) x que satisface f(x) = y.
  """


# Ejemplo de uso
x_min = -3
x_max = 3
n_puntos = 100

# Valores de x y y para la función original
x_original = []
y_original = []
for i in range(n_puntos):
  x = x_min + (i * (x_max - x_min)) / (n_puntos - 1)
  y = funcion_original(x)
  x_original.append(x)
  y_original.append(y)

# Valores de x y y para la función inversa
x_inversa = []
y_inversa = []
for i in range(n_puntos):
  y = i * (0.1)  # Ajustar el rango de y según la función inversa
  x = funcion_inversa(y)
  x_inversa.append(x)
  y_inversa.append(y)

# Graficar funciones original e inversa
plt.plot(x_original, y_original, label="Función original f(x)")
plt.plot(x_inversa, y_inversa, label="Función inversa f^-1(y)")

# Ajustar el rango de los ejes
plt.xlim([x_min, x_max])
plt.ylim([-1, y_original[-1] + 1])  # Ajustar rango de y según la gráfica

# Etiquetas y título
plt.xlabel("Eje X (x)")
plt.ylabel("Eje Y (y)")
plt.title("Gráfica de Función Original e Inversa")

# Leyenda y grilla
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```
