## üîç Qu√© son las funciones matem√°ticas?

En el mundo de las matem√°ticas, las funciones son conceptos fundamentales que describen la relaci√≥n entre dos conjuntos de n√∫meros. Se las puede visualizar como reglas o mecanismos que asignan un valor de salida (llamado imagen) a cada valor de entrada (llamado dominio).

![Texto alternativo](https://datascientest.com/es/wp-content/uploads/sites/7/2022/06/ejemplo_fonciones-1-1024x690.png)


### üìòEjemplo

Imaginemos una m√°quina expendedora:

* Entrada: Insertas una moneda (valor de entrada).
Funci√≥n: La m√°quina interpreta la moneda y determina el producto correspondiente (regla o mecanismo).
Salida: Recibes el producto seleccionado (imagen o valor de salida).
Caracter√≠sticas esenciales de las funciones:

* Dominio: El conjunto de todos los valores de entrada v√°lidos para la funci√≥n. Se representa con la letra D.

* Codominio: El conjunto que contiene todos los valores de salida posibles de la funci√≥n. Se representa con la letra C.

* Imagen: El subconjunto del codominio que contiene todos los valores de salida efectivamente asignados por la funci√≥n. Se representa con la letra I.

* Unicidad: Para cada valor de entrada en el dominio, solo hay un valor de salida correspondiente en la imagen.

## üßÆRepresentaciones comunes de funciones:

* F√≥rmulas matem√°ticas: La forma m√°s com√∫n de expresar una funci√≥n es mediante una f√≥rmula que relaciona la variable de entrada (x) con la variable de salida (y). Por ejemplo, la funci√≥n lineal y = mx + b, donde m y b son constantes.

* Tablas de valores: Se muestran pares de valores de entrada y salida, permitiendo observar la relaci√≥n entre ellos.

* Gr√°ficas: Representan visualmente la relaci√≥n entre la variable de entrada (eje x) y la variable de salida (eje y). La curva de la gr√°fica indica c√≥mo cambia el valor de salida a medida que var√≠a la entrada.

```python
import math

def funcion(x):
  """
  Funci√≥n f(x) = x^2 + 1.

  Args:
    x: Valor de entrada (variable independiente).

  Returns:
    Valor de salida (imagen) de la funci√≥n f(x) para el valor de entrada x.
  """
  return x**2 + 1

def imagen(x_min, x_max):
  """
  Calcula la imagen de la funci√≥n f(x) en el intervalo [x_min, x_max].

  Args:
    x_min: Limite inferior del intervalo.
    x_max: Limite superior del intervalo.

  Returns:
    Lista de valores de imagen (f(x)) para cada valor de x en el intervalo.
  """
  valores_imagen = []
  for x in range(x_min, x_max + 1):
    valor_imagen = funcion(x)
    valores_imagen.append(valor_imagen)
  return valores_imagen

def preimagen(y):
  """
  Calcula la preimagen (valores de x) para un valor de imagen y de la funci√≥n f(x).

  Args:
    y: Valor de imagen (f(x)) para el que se busca la preimagen.

  Returns:
    Lista de valores de x que satisfacen f(x) = y.
  """
  valores_preimagen = []
  for x in range(-10, 11):  # Ajustar el rango seg√∫n la funci√≥n
    valor_imagen = funcion(x)
    if valor_imagen == y:
      valores_preimagen.append(x)
  return valores_preimagen

# Ejemplo de uso
x_min = -3
x_max = 3
y = 5

valores_imagen = imagen(x_min, x_max)
valores_preimagen = preimagen(y)

print("Imagen de f(x) en [", x_min, ",", x_max, "]:", valores_imagen)
print("Preimagen de f(x) para y =", y, ":", valores_preimagen)
```

```python
import math
import matplotlib.pyplot as plt

def funcion(x):
  """
  Funci√≥n f(x) = x^2 + 1.

  Args:
    x: Valor de entrada (variable independiente).

  Returns:
    Valor de salida (imagen) de la funci√≥n f(x) para el valor de entrada x.
  """
  return x**2 + 1

def imagen(x_min, x_max, n_puntos):
  """
  Calcula y grafica la imagen de la funci√≥n f(x) en el intervalo [x_min, x_max].

  Args:
    x_min: Limite inferior del intervalo.
    x_max: Limite superior del intervalo.
    n_puntos: Numero de puntos para calcular la imagen.

  Returns:
    Listas de valores de x (eje x) y y (eje y) para la imagen de la funci√≥n.
  """
  x_valores = []
  y_valores = []
  for i in range(n_puntos):
    x = x_min + (i * (x_max - x_min)) / (n_puntos - 1)
    y = funcion(x)
    x_valores.append(x)
    y_valores.append(y)
  return x_valores, y_valores

def preimagen(y, x_min, x_max, n_puntos):
  """
  Calcula y grafica la preimagen (valores de x) para un valor de imagen y de la funci√≥n f(x).

  Args:
    y: Valor de imagen (f(x)) para el que se busca la preimagen.
    x_min: Limite inferior del intervalo de b√∫squeda.
    x_max: Limite superior del intervalo de b√∫squeda.
    n_puntos: Numero de puntos para calcular la preimagen.

  Returns:
    Lista de valores de x (eje x) que satisfacen f(x) = y.
  """
  x_preimagen = []
  for i in range(n_puntos):
    x = x_min + (i * (x_max - x_min)) / (n_puntos - 1)
    valor_imagen = funcion(x)
    if abs(valor_imagen - y) < 1e-5:  # Tolerancia para la aproximaci√≥n
      x_preimagen.append(x)
  return x_preimagen

# Ejemplo de uso
x_min = -3
x_max = 3
n_puntos = 100
y = 5

x_imagen, y_imagen = imagen(x_min, x_max, n_puntos)
x_preimagen = preimagen(y, x_min, x_max, n_puntos)

# Graficar imagen
plt.plot(x_imagen, y_imagen, label="Imagen f(x)")
plt.axhline(y=y, color="red", linestyle="--", label=f"Imagen fija y = {y}")

# Graficar preimagen (si existe)
if x_preimagen:
  for x in x_preimagen:
    plt.axvline(x=x, color="blue", linestyle="--", label=f"Preimagen x = {x}")

plt.xlim([x_min, x_max])
plt.ylim([0, y_imagen[-1] + 1])  # Ajustar rango de y seg√∫n la gr√°fica
plt.xlabel("Eje X (x)")
plt.ylabel("Eje Y (y)")
plt.title("Imagen y Preimagen de f(x) = x^2 + 1")
plt.legend()
plt.grid(True)
plt.show()
```

### üîÑ Definici√≥n de Funci√≥n
Una funci√≥n $(f$) de un conjunto $(A$) a un conjunto $(B$) es una relaci√≥n que asigna a cada elemento de $(A$) exactamente un elemento de $(B$). Se denota como $(f: A \to B$).

### üìâFunci√≥n original:

Una funci√≥n original, tambi√©n conocida como funci√≥n directa o simplemente funci√≥n, se define como una relaci√≥n que asigna a cada elemento del dominio (conjunto de valores de entrada v√°lidos) un √∫nico elemento del codominio (conjunto de valores de salida posibles).

### Representaci√≥n de una funci√≥n original:

F√≥rmulas matem√°ticas: La forma m√°s com√∫n de expresar una funci√≥n original es mediante una f√≥rmula que relaciona la variable de entrada (x) con la variable de salida (y).

Por ejemplo, la funci√≥n lineal y = mx + b, donde m y b son constantes.

Tablas de valores: Se muestran pares de valores de entrada (x) y salida (y), permitiendo observar la correspondencia entre ellos.

Gr√°ficas: Representan visualmente la relaci√≥n entre la variable de entrada (eje x) y la variable de salida (eje y).

 La curva de la gr√°fica indica c√≥mo cambia el valor de salida a medida que var√≠a la entrada.

###üîéPropiedades de una funci√≥n original:

Unicidad: Para cada valor de entrada en el dominio, solo hay un valor de salida correspondiente en la imagen (el subconjunto del codominio que contiene los valores de salida efectivamente asignados).

No necesariamente invertible: No todas las funciones originales tienen una funci√≥n inversa. Una funci√≥n solo es invertible si es mon√≥tona y continua en su inverso.

### Ejemplos de funciones originales:

Funci√≥n lineal: y = mx + b

Funci√≥n cuadr√°tica: y = ax^2 + bx + c

Funci√≥n exponencial: y = a^x (donde a > 0, a ‚â† 1)

Funci√≥n logar√≠tmica: y = log_a(x) (donde a > 0, a ‚â† 1)

Funci√≥n trigonom√©trica: y = sin(x), y = cos(x), y = tan(x)

##üìíFunci√≥n inversa:

La funci√≥n inversa, tambi√©n conocida como funci√≥n rec√≠proca, es la funci√≥n que "deshace" lo que hace la funci√≥n original. En otras palabras, si f(x) = y, entonces la funci√≥n inversa f^(-1)(y) = x, donde x es el valor de entrada original que produce la salida y.

###Representaci√≥n de una funci√≥n inversa:

F√≥rmula: Si la funci√≥n original tiene una f√≥rmula expl√≠cita, la funci√≥n inversa puede obtenerse despejando x en y = f(x).

Gr√°fica: La gr√°fica de la funci√≥n inversa es la sim√©trica de la gr√°fica de la funci√≥n original respecto a la bisectriz de los cuadrantes.
Propiedades de una funci√≥n inversa:

Inversa de la original: Si f(f^(-1)(y)) = y para todo y en el dominio de la funci√≥n inversa, y f^(-1)(f(x)) = x para todo x en el dominio de la funci√≥n original, entonces f y f^(-1) son funciones inversas entre s√≠.

Monotonicidad: La funci√≥n inversa es mon√≥tona (creciente o decreciente) si y solo si la funci√≥n original es mon√≥tona estrictamente creciente o decreciente.

###Ejemplos de funciones inversas:

Funci√≥n lineal inversa: y = (1/m) * (x - b)

Funci√≥n cuadr√°tica inversa: x = (-b ¬± sqrt(b^2 - 4ac)) / 2a

Funci√≥n exponencial inversa: x = log_a(y) (donde a > 0, a ‚â† 1)

Funci√≥n logar√≠tmica inversa: x = e^y

Funci√≥n trigonom√©trica inversa: x = arcsin(y), x = arccos(y), x = arctan(y)

##üë®‚ÄçüíªRelaci√≥n entre funci√≥n original e inversa:

La gr√°fica de la funci√≥n inversa se obtiene tomando la gr√°fica de la funci√≥n original y reflej√°ndola respecto a la bisectriz de los cuadrantes.

Las funciones inversas solo existen para funciones originales que son mon√≥tonas estrictamente crecientes o decrecientes en todo su dominio.

Si f(x) = y, entonces f^(-1)(y) = x, donde x es el valor de entrada original que produce la salida y.




```python
import math
import matplotlib.pyplot as plt

def funcion_original(x):
  """
  Funci√≥n original f(x) = x^2 + 1.

  Args:
    x: Valor de entrada (variable independiente).

  Returns:
    Valor de salida (imagen) de la funci√≥n f(x) para el valor de entrada x.
  """
  return x**2 + 1

def funcion_inversa(y):
  """
  Funci√≥n inversa f^-1(y) de f(x) = x^2 + 1.

  Args:
    y: Valor de salida (imagen) para el que se busca la preimagen.

  Returns:
    Valor de entrada (preimagen) x que satisface f(x) = y.
  """


# Ejemplo de uso
x_min = -3
x_max = 3
n_puntos = 100

# Valores de x y y para la funci√≥n original
x_original = []
y_original = []
for i in range(n_puntos):
  x = x_min + (i * (x_max - x_min)) / (n_puntos - 1)
  y = funcion_original(x)
  x_original.append(x)
  y_original.append(y)

# Valores de x y y para la funci√≥n inversa
x_inversa = []
y_inversa = []
for i in range(n_puntos):
  y = i * (0.1)  # Ajustar el rango de y seg√∫n la funci√≥n inversa
  x = funcion_inversa(y)
  x_inversa.append(x)
  y_inversa.append(y)

# Graficar funciones original e inversa
plt.plot(x_original, y_original, label="Funci√≥n original f(x)")
plt.plot(x_inversa, y_inversa, label="Funci√≥n inversa f^-1(y)")

# Ajustar el rango de los ejes
plt.xlim([x_min, x_max])
plt.ylim([-1, y_original[-1] + 1])  # Ajustar rango de y seg√∫n la gr√°fica

# Etiquetas y t√≠tulo
plt.xlabel("Eje X (x)")
plt.ylabel("Eje Y (y)")
plt.title("Gr√°fica de Funci√≥n Original e Inversa")

# Leyenda y grilla
plt.legend()
plt.grid(True)

# Mostrar la gr√°fica
plt.show()
```
